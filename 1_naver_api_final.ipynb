```python
# 최종 코드
# 블로그 검색 api
# 한 번에 가져오는 개수인 display는 최대 100까지 가능
# start변수 값을 바꿔야 다른 값을 불러옴, 최대 1000까지
# display=100이면 start변수 값을 1, 101, 201, ..., 1000으로 바꿔주면서 해야함
# start는 최대 1000까지 제공하기 때문에, start=1일때랑 start=1001일때랑 결과가 같음


import os
import sys
import urllib.request

client_id = ""  # 본인 클라이언트 아이디 넣기
client_secret = ""  # 본인 클라이언트 시크릿 넣기
encText = urllib.parse.quote("디저트")

url = "https://openapi.naver.com/v1/search/blog?query="+encText+"&display=100"+"&start=1000" # json 결과
   
file=open("./naver_api_dessert11.json","w+",encoding='UTF8')  # 파일로 저장하기
request = urllib.request.Request(url)
request.add_header("X-Naver-Client-Id",client_id)
request.add_header("X-Naver-Client-Secret",client_secret)
response = urllib.request.urlopen(request)
rescode = response.getcode()
if(rescode==200):
    response_body = response.read()
    print(response_body.decode('utf-8'))
    file.write(response_body.decode('utf-8'))
    file.close()
else:
    print("Error Code:" + rescode)
```
